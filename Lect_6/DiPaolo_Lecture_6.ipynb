# Lecture 6 - Ising model

One of the most researched model in statistical physics is the Ising model, which in one dimension (1D) is exactly resoluble. 
The Ising model is a model of a magnet and it relies on the fact that the magnetism of a bulk material is made up of the combined magnetic dipole moments of many atomic spins within the material. The model postulates a lattice with a magnetic dipole/spin on each site.
In this model these spins assume the simplest form possible, which consists of scalar variables $s_i$ which can take only two values $\pm 1$, representing up-pointing or down-pointing dipoles of unit magnitude.
The relationship between the total angular momentum $\mathbf{j}$ and the magnetic dipole momentum $\mathbf{\mu}$ is 

$$
\mathbf{\mu} = g\dfrac{q_e}{2mc} \mathbf{j} \hspace{0.4cm} \text{and} \hspace{0.4cm} j^2 = \hbar^2 J(J+1)
$$
where $g$ is the Landé factor.
Thus leading to
$$
\mathbf{\mu}^2 = \left(g\dfrac{q_e}{2mc}\right)^2 \hbar^2 J(J+1) = g^2 \mu_B^2 J(J+1)
$$
where $\mu_B = \frac{q_e \hbar}{2mc}$ is the Bohr magnetron.

The component of the magnetic moment along the z-axis $\mu_z$ is given by
$$
\mu_z = g \mu_B m \hspace{0.5cm} \text{where}\hspace{0.3cm} m=-J, -J+1, \dots, J-1,J
$$
From a lot of experimental and theoretical studies a magnetic moment $j=\frac{1}{2}$ for many ferromagnetic materials has been shown, thus leading to a more simple relationship
$$
\mu = g^2 \mu_B^2 s(s+1) 
$$
where $s$ is equal to $\frac{1}{2}$ and $\mu_z$ can assume only two values that correspond to $m=\pm J = \pm \frac{1}{2}$. With the value $s=\frac{1}{2}$ only two orientations are possible for each lattice site. 

This is the classical Ising model where the energy is measured as (with the assumption of $\mu_B=1$ and $k_B=1$):
$$
H = 
-J\sum_{i=1}^N s_i s_{i+1}
-\frac{h}{2}\sum_{i=1}^N (s_i + s_{i+1}) 
$$
where $\{s_i\}_{i=1,\dots,N}$ are the spin variables (they can assume the values $\pm 1$), $h$ is the applied external field and $J$ is the tendency of electrons to align themselves parallel/anti-parallel due to the exchange interaction.

The partition function for $N$ spins, $Z$, can be exactly computed and it turns out to be:

$$
Z=\sum_{\{ s_i \}} e^{ -\beta H} = \lambda_1^N + \lambda_2^N
$$

where
$\lambda_{1,2} = 
e^{\beta J} 
\cosh (\beta h) \pm \left[ e^{2\beta J} \cosh^2 (\beta h) - 2\sinh (2\beta J) \right]^{1/2}$, 
and where $\beta=\frac{1}{k_B T} = \frac{1}{T}$.
Once the partition function has been computed, the Helmholtz free energy is readly obtained as $A = -k_B T \ln Z$. All the thermodynamic quantities of interest can be deduced from $A$ or, equivalently, from $Z$.

All the thermodynamic quantities of interest can be deduced from $A$ or, equivalently, from $Z$:

- the **Internal Energy** per spin (with $h=0$):

$$
\frac{U(N,T)}{N} = \frac{\langle H \rangle}{N}
= \frac{1}{N} \sum_{\{ s_i \}} H \frac{\exp \left( -\beta H\right)}{Z}
= -\frac{1}{N}\frac{\partial \ln Z}{\partial \beta}
\underset{h=0}= -J \tanh (\beta J) \Biggl( \frac{1 + \tanh^{N-2} (\beta J)}{1+\tanh^N (\beta J)} \Biggl)
$$

- the **Heat Capacity** (with $h=0$):

$$
C(N,T) = \frac{\partial U(N,T)}{\partial T}
= \frac{\partial \beta}{\partial T}\frac{\partial U}{\partial \beta}
= -k_B\beta^2 \frac{\partial}{\partial \beta} \left[ \sum_{\{ s_i \}} H \frac{\exp \left( -\beta H\right)}{Z} \right] = \\
= -k_B\beta^2 \left[ -\sum_{\{ s_i \}} H^2 \frac{\exp \left( -\beta H\right)}{Z} + \sum_{\{ s_i \}} H \exp \left( -\beta H\right)\frac{\partial Z^{-1}}{\partial \beta} \right] = \\
= k_B\beta^2 \left\{ \sum_{\{ s_i \}} H^2 \frac{\exp \left( -\beta H\right)}{Z} + \left[ \sum_{\{ s_i \}} H \frac{\exp \left( -\beta H\right)}{Z}\right]\frac{1}{Z}\frac{\partial Z}{\partial \beta} \right\} =\\
= k_B\beta^2 \left\{ \sum_{\{ s_i \}} H^2 \frac{\exp \left( -\beta H\right)}{Z} - \left[ \sum_{\{ s_i \}} H \frac{\exp \left( -\beta H\right)}{Z}\right]^2 \right\} =
k_B\beta^2 \left(\langle H^2 \rangle -\langle H \rangle^2 \right)
$$

$$
\frac{C(N,T)}{N} = \frac{1}{N}\frac{\partial U(N,T)}{\partial T} =\\
\underset{h=0}= 
k_B (\beta J)^2 \left\{ 1+\tanh^N (\beta J)+(N-1)\tanh^2 (\beta J)+(N-1)\frac{\tanh^{N-2} (\beta J)}{1+\tanh^N (\beta J)}
- N \left[ \frac{\tanh (\beta J) + \tanh^{N-1} (\beta J)}{1+\tanh^N (\beta J)} \right]^2 \right\}
$$

- The **magnetization** (with $h \not= 0$):

Note that with $h=0$ the magnetization $M(N,T,h=0)=\langle \sum_{i=1,N}s_i \rangle =0$ because in 1D we cannot have a (ferromagnetic) phase transition in absence of an external magnetic field.

$$
M(N,T,h) = -\frac{\partial A}{\partial h} = k_B T \frac{\partial \ln Z}{\partial h} =
\frac{1}{\beta} \frac{1}{Z}\frac{\partial}{\partial h}\sum_{\{ s_i \}} \exp \left( -\beta H\right) = \sum_{\{ s_i \}} \left[ \sum_{i=1,N}s_i\right] \frac{\exp \left( -\beta H\right)}{Z} = \left\langle \sum_{i=1,N}s_i \right\rangle
$$

The exact formula from $M(N,T,h) = k_B T \dfrac{\partial \ln Z}{\partial h}$ is:
$$
\frac{M(N,T,h)}{N}=\frac{e^{\beta J}\; \sinh(\beta h)}{Z} \; \Bigl[\lambda_1^{N-1}\cdot \Bigl(1+\frac{e^{\beta J}\; \cosh(\beta h)}{\sqrt{e^{2\beta J} \cosh^2 (\beta h) - 2\sinh (2\beta J)}}\Bigr) + \lambda_2^{N-1}\cdot \Bigl(1-\frac{e^{\beta J}\; \cosh(\beta h)}{\sqrt{e^{2\beta J} \cosh^2 (\beta h) - 2\sinh (2\beta J)}}\Bigr)\Bigr]
$$

- The **magnetic susceptibility** (with $h=0$):

$$
\chi(N,T) = \frac{\partial M}{\partial h} = 
\frac{\partial}{\partial h}\sum_{\{ s_i \}} \left[ \sum_{i=1,N}s_i\right] \frac{\exp \left( -\beta H\right)}{Z} =\\
= \beta\sum_{\{ s_i \}} \left[ \sum_{i=1,N}s_i\right]^2 \frac{\exp \left( -\beta H\right)}{Z}
-\sum_{\{ s_i \}} \left[ \sum_{i=1,N}s_i\right] \frac{\exp \left( -\beta H\right)}{Z}
\frac{1}{Z}\frac{\partial Z}{\partial h} = 
\beta \left[ \left\langle \left( \sum_{i=1,N}s_i \right)^2 \right\rangle - \left\langle \sum_{i=1,N}s_i \right\rangle^2 \right] = \\
\underset{h=0}= \beta \left[ \left\langle \left( \sum_{i=1,N}s_i \right)^2 \right\rangle \right] = \beta e^{2\beta J} \Biggl( \frac{1-\tanh^N (\beta J)}{1+\tanh^N (\beta J)} \Biggl)
$$

## Solution
The accessible states of a Ising system are the different sets of values that the spins can take. Since each spin can take two values, there are a total of $2^N$ states for a lattice with $N$ spins. The partition function of the model is the sum

$$
Z = \sum_{\{s_i\}} e^{-\beta H} = \sum_{s_1 = \pm 1}\sum_{s_2 = \pm 1}\dots \sum_{s_N = \pm 1} \exp \left[\beta J \sum_{\langle ij\rangle}s_is_j +\beta B \sum_i s_i\right]
$$

This is the key quantity that we want to compute via Monte Carlo methods for a finite-size Ising system. 

Before simulation, we have to define the environment and the starting configuration:
- Periodic Boundary Conditions 

    Used to ensure all spins have the same number of neighbors, simulating an infinite lattice and removing "edge effects". 
- Initial States ($\mu$):

    The ground State ($T=0$), where all spins are aligned (all "up" or all "down") and the infinite-temperature state ($T = \infty$)
where the spins are oriented randomly in an uncorrelated fashion.

### Metropolis algorithm 
In the application of M(RT)$^2$ algorithm to this system we always have to define a trial move. So we propose a change from the state $\mu$ to the new state $\nu$ 
- Pick a single spin $k$ at random from the lattice.
- Flip the spin to create the trial state $\nu$.
- Because the selection is random, the trial transition probability is symmetric: $T(\nu|\mu) = T(\mu|\nu)$.

Then we decide whether to accept the new state $\nu$ based on the Metropolis algorithm, so the probability of acceptance is given by

$$
A(\nu|\mu) = \min\left[1, \frac{p(\nu)}{p(\mu)}\right] = \min\left[1, e^{-\beta(E_\nu - E_\mu)}\right]
$$

The idea is that 
- if $\Delta E \leq 0$, so the energy does not increase we always accept
- if $\Delta E > 0$, we accept only with a probability of $e^{-\beta \Delta E}$.

We have to note that in the implementation - due to efficiency issues - instead of calculating the total Hamiltonian, we only calculate the local change in energy for the flipped spin $k$:

$$
\Delta E = E_\nu - E_\mu = -J \sum_{\langle i,j \rangle} s_i^\nu s_j^\nu + J \sum_{\langle i,j \rangle} s_i^\mu s_j^\mu
$$

For a single spin flip at site $k$, this simplifies to looking only at the close neighbors ($nn$)
$$
\Delta E = 2 J s_k \sum_{i \in nn(k)} s_i
$$

In practice we evaluate the acceptance ratio $A(n|\mu)$ using our value of $E_n-E_\mu$, we produce a random number $r \in [0,1)$ 
- if $r < A(n|\mu) \rightarrow $ we flip the spin
- if $r > A(n|\mu) \rightarrow $ we do not change the spin 

### Gibbs method
This method also known as the heat-bath algorithm, is conceptually the simplest of the Markov chain sampling methods and it provides an alternative approach to the Metropolis algorithm for simulating the Ising model. 

It can be seen as a particular case of the Metropolis algorithm where the proposed move is always accepted; we sample the new spin value directly from its local conditional distribution.

We start by selecting the initial values for all spins $s_i$ according to an initial distribution that we have to equilibrate, subsequently we visit each spin $s_i$ repeatedly, either in a fixed order or by selecting one at random. Once a spin $s_i$ is visited, its new value is chosen independently of its current value, based strictly on its nearest neighbors (n.n.) and so determined by the conditional distribution defined by the surrounding system:
$$
p(s'_k | \{s_j : j \neq k\}) = \frac{p(s_1, \dots, s'_k, \dots, s_N)}{p(s_1, \dots, s_k=+1, \dots, s_N) + p(s_1, \dots, s_k=-1, \dots, s_N)}
$$
In terms of the energy change $\Delta E$ and interaction strength $J$, the probability of setting a spin to $\pm 1$ is:
$$
p(s'_k = \pm 1 | \{s_j : j \neq k\}) = \frac{1}{1 + \exp\left(\mp 2\beta J \sum_{i \in n.n.} s_i\right)}
$$
Once this computation is done for each spin site we performed a Gibbs step and after multiple steps the system will follow the typical Boltzmann probability distribution
$$
P(\{s_i\}) \propto e^{-\beta E(\{s_i\})}
$$

### Equilibration
To compute meaningful averages, we must remove an initial part of the Monte Carlo (MC): the so-called equilibration.

In a well-equilibrated simulation, the system reaches a stationary distribution. At this point, the probability $P(s)$ of finding the system in a specific state $s$ is directly proportional to its Boltzmann weight:
$$
P(s) \propto e^{-\beta E(s)}
$$
Where $\beta = \frac{1}{k_B T}$ and $E(s)$ is the energy of the state. Until this condition is met, any collected data will be biased by the initial conditions.

Since we cannot look at the probability distribution directly, we monitor collective variables (e.g internal energy) over the course of the simulation. Once equilibrated, the variable should move around a stable average.

When the system reaches the equilibrium we can measure the average of the quantity $G(\mu)$ that we are interested in as

$$
\langle G \rangle \approx \dfrac{1}{N-N_\text{eq}} \sum_{i=N_\text{eq} +1}^N G(\mu_i)
$$

$N_\text{eq}$ being the number of Monte Carlo steps of equilibration and $N$ being the total number of steps.

In this code we performed a temperature sweep starting from a high-temperature state ($T=2$ K) and cooling down to a lower temperature state ($T=0.5$ K) with $n = 30$ sampling points.

The code performs $N_{eq}$ Monte Carlo steps needed as equilibration steps, the value was determined by the two short initial simulations. These steps were discarded before taking measurements at each new temperature, thus leading the lattice to reach thermal equilibrium at the new $T$.

To estimate statistical uncertainties and avoid the correlation between consecutive Monte Carlo steps, the `NSL_simulator` code employs the data blocking technique, which can be summarized as:

1. For each block $i$, the system evolves for a fixed number of steps (`get_nsteps`).
2. During these steps, the code calls `SYS.measure()` to record instantaneous values.
3. At the end of the block, `SYS.averages(i+1)` calculates the mean for that block and updates the cumulative global average and its error (standard deviation of the mean).



```python
points = 100
T = np.linspace(0.5, 2.0, num=points)
beta = 1/T
J = 1.0
Ns = 50
th = np.tanh(J/T)
thN = th**Ns
ch = 1/th

fig, axs = plt.subplots(2, 2, figsize=(14, 10))
# =====================================================
# INTERNAL ENERGY
e = -J*(th + ch*thN)/(1 + thN)

axs[0,0].plot(T, e, color=c_theory, linewidth=1.2, label='Theory')

data_m = np.loadtxt("OUTPUT/METROPOLIS/LONGER/filtered_total_energy.dat")
axs[0,0].errorbar(data_m[:,0], data_m[:,1], yerr=data_m[:,3],
                  fmt='o', color=c_metro, markersize=4, label='Metropolis')

data_g = np.loadtxt("OUTPUT/GIBBS/LONGER/filtered_total_energy.dat")
axs[0,0].errorbar(data_g[:,0], data_g[:,1], yerr=data_g[:,3],
                  fmt='s', color=c_gibbs, markersize=4, label='Gibbs')

axs[0,0].set_xlabel("Temperature")
axs[0,0].set_ylabel("Internal energy (h = 0)")
axs[0,0].grid(True, linestyle='--', alpha=0.6)
axs[0,0].legend()

axins = inset_axes(
    axs[0,0],
    width="40%",
    height="40%",
    loc="lower right",
    bbox_to_anchor=(0, 0.08, 1, 1),
    bbox_transform=axs[0,0].transAxes
)

e_interp_m = np.interp(data_m[:,0], T, e)
diff_m = np.abs(data_m[:,1] - e_interp_m)

e_interp_g = np.interp(data_g[:,0], T, e)
diff_g = np.abs(data_g[:,1] - e_interp_g)

axins.errorbar(data_m[:,0], diff_m, yerr=data_m[:,3],
               fmt='o', color=c_metro, markersize=3)
axins.errorbar(data_g[:,0], diff_g, yerr=data_g[:,3],
               fmt='s', color=c_gibbs, markersize=3)

axins.axhline(0, color='black', linestyle='-.')
axins.set_ylabel("Error", fontsize=9)
axins.grid(True, linestyle='--', alpha=0.5)


# =====================================================
# SPECIFIC HEAT
heat = ((beta*J)**2)*(
    (1 + thN + (Ns-1)*th**2 + (Ns-1)*(ch**2)*thN)/(1 + thN)
    - Ns*((th + ch*thN)/(1 + thN))**2
)

axs[0,1].plot(T, heat, color=c_theory, linewidth=1.2)

data_m = np.loadtxt("OUTPUT/METROPOLIS/LONGER/specific_heat.dat")
axs[0,1].errorbar(data_m[:,0], data_m[:,1], yerr=data_m[:,2],
                  fmt='o', color=c_metro, markersize=4)

data_g = np.loadtxt("OUTPUT/GIBBS/LONGER/specific_heat.dat")
axs[0,1].errorbar(data_g[:,0], data_g[:,1], yerr=data_g[:,2],
                  fmt='s', color=c_gibbs, markersize=4)

axs[0,1].set_xlabel("Temperature")
axs[0,1].set_ylabel("Specific heat (h = 0)")
axs[0,1].grid(True, linestyle='--', alpha=0.6)

axins = inset_axes(
    axs[0,1],
    width="40%",
    height="40%",
    loc="lower center",
    bbox_to_anchor=(0, 0.08, 0.9, 1),
    bbox_transform=axs[0,1].transAxes
)

heat_interp_m = np.interp(data_m[:,0], T, heat)
diff_m = np.abs(data_m[:,1] - heat_interp_m)

heat_interp_g = np.interp(data_g[:,0], T, heat)
diff_g = np.abs(data_g[:,1] - heat_interp_g)

axins.errorbar(data_m[:,0], diff_m, yerr=data_m[:,2],
               fmt='o', color=c_metro, markersize=3)
axins.errorbar(data_g[:,0], diff_g, yerr=data_g[:,2],
               fmt='s', color=c_gibbs, markersize=3)

axins.axhline(0, color='black', linestyle='-.')
axins.set_ylabel("Error", fontsize=9)
axins.grid(True, linestyle='--', alpha=0.5)


# =====================================================
# MAGNETIZATION
h = 0.02
b = 1/T

l1 = np.exp(b*J)*np.cosh(b*h) + np.sqrt(np.exp(2*b*J)*np.cosh(b*h)**2 - 2*np.sinh(2*b*J))
l2 = np.exp(b*J)*np.cosh(b*h) - np.sqrt(np.exp(2*b*J)*np.cosh(b*h)**2 - 2*np.sinh(2*b*J))
Z = l1**Ns + l2**Ns

M = (np.exp(b*J)*np.sinh(b*h) *
     ((l1**(Ns-1))*(1 + np.exp(b*J)*np.cosh(b*h) /
      np.sqrt(np.exp(2*b*J)*np.cosh(b*h)**2 - 2*np.sinh(2*b*J)))
    + (l2**(Ns-1))*(1 - np.exp(b*J)*np.cosh(b*h) /
      np.sqrt(np.exp(2*b*J)*np.cosh(b*h)**2 - 2*np.sinh(2*b*J)))
    ) / Z)

axs[1,0].plot(T, M, color=c_theory, linewidth=1.2)

data_m = np.loadtxt("OUTPUT/METROPOLIS/LONGER/magnetization.dat")
axs[1,0].errorbar(data_m[:,0], data_m[:,1], yerr=data_m[:,2],
                  fmt='o', color=c_metro, markersize=4)

data_g = np.loadtxt("OUTPUT/GIBBS/LONGER/magnetization.dat")
axs[1,0].errorbar(data_g[:,0], data_g[:,1], yerr=data_g[:,2],
                  fmt='s', color=c_gibbs, markersize=4)

axs[1,0].set_xlabel("Temperature")
axs[1,0].set_ylabel("Magnetization (h = 0.02)")
axs[1,0].grid(True, linestyle='--', alpha=0.6)

axins = inset_axes(axs[1,0], width="40%", height="40%", loc="upper right")

M_interp_m = np.interp(data_m[:,0], T, M)
diff_m = np.abs(data_m[:,1] - M_interp_m)

M_interp_g = np.interp(data_g[:,0], T, M)
diff_g = np.abs(data_g[:,1] - M_interp_g)

axins.errorbar(data_m[:,0], diff_m, yerr=data_m[:,2],
               fmt='o', color=c_metro, markersize=3)
axins.errorbar(data_g[:,0], diff_g, yerr=data_g[:,2],
               fmt='s', color=c_gibbs, markersize=3)

axins.axhline(0, color='black', linestyle='-.')
axins.set_ylabel("Error", fontsize=9)
axins.grid(True, linestyle='--', alpha=0.5)


# =====================================================
# SUSCEPTIBILITY
X = beta*np.exp(2*beta*J)*(1 - thN)/(1 + thN)

axs[1,1].plot(T, X, color=c_theory, linewidth=1.2)

data_m = np.loadtxt("OUTPUT/METROPOLIS/LONGER/susceptibility.dat")
axs[1,1].errorbar(data_m[:,0], data_m[:,1], yerr=data_m[:,2],
                  fmt='o', color=c_metro, markersize=4, label='Metropolis')

data_g = np.loadtxt("OUTPUT/GIBBS/LONGER/susceptibility.dat")
axs[1,1].errorbar(data_g[:,0], data_g[:,1], yerr=data_g[:,2],
                  fmt='s', color=c_gibbs, markersize=4, label='Gibbs')

axs[1,1].set_xlabel("Temperature")
axs[1,1].set_ylabel("Susceptibility (h = 0)")
axs[1,1].grid(True, linestyle='--', alpha=0.6)
axs[1,1].legend()

axins = inset_axes(axs[1,1], width="40%", height="40%", loc="upper right")

X_interp_m = np.interp(data_m[:,0], T, X)
diff_m = np.abs(data_m[:,1] - X_interp_m)

X_interp_g = np.interp(data_g[:,0], T, X)
diff_g = np.abs(data_g[:,1] - X_interp_g)

axins.errorbar(data_m[:,0], diff_m, yerr=data_m[:,2],
               fmt='o', color=c_metro, markersize=3)
axins.errorbar(data_g[:,0], diff_g, yerr=data_g[:,2],
               fmt='s', color=c_gibbs, markersize=3)

axins.axhline(0, color='black', linestyle='-.')
axins.set_ylabel("Error", fontsize=9)
axins.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

We note that the results are compatible with the theoretical curves, but in the case of specific heat, the sampling at low temperatures proves to be not accurate. The moves of the Metropolis algorithm are accepted with a probability dominated by the Boltzmann weight, which decreases exponentially as the temperature increases. This implies that most of the moves that would lead the system to a higher energy state are rejected, leaving the system frozen in a specific energy configuration. Consequently, the phase space is not sampled correctly.
