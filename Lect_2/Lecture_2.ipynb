```python
import matplotlib.pyplot as plt
import numpy as np

file1 = "Exercise_2_1/output_average.data"
file2 = "Exercise_2_1/output_impsampling.data"
x1, y1, error1 = np.loadtxt(file1, usecols=(0, 1, 2), delimiter='\t', unpack=True)
x2, y2, error2 = np.loadtxt(file2, usecols=(0, 1, 2), delimiter='\t', unpack=True)

fig, axes = plt.subplots(1, 2, figsize=(12, 5))
viridis = plt.cm.viridis(np.linspace(0, 1, 4))

axes[0].errorbar(x1, y1, yerr=error1, fmt='o', color=viridis[0], capsize=0, elinewidth=2, alpha=0.4)
axes[0].set_xlabel('Number of throws')
axes[0].set_ylabel('Average')
axes[0].grid(True)

axes[1].errorbar(x2, y2, yerr=error2, fmt='o', color=viridis[1], capsize=0, elinewidth=2, alpha=0.4)
axes[1].set_xlabel('Number of throws')
axes[1].set_ylabel('Average - Importance sampling')
axes[1].grid(True)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

# Funzione da fittare
def fit_func(x, k):
    return k * np.sqrt(x)

# Caricamento dati
file1 = "Exercise_2_2/output_RW_discrete_1.000000.data"
file2 = "Exercise_2_2/output_RW_discrete_3.000000.data"
file3 = "Exercise_2_2/output_RW_discrete_10.000000.data"

x1, y1, error1 = np.loadtxt(file1, usecols=(0, 1, 2), delimiter='\t', unpack=True)
x2, y2, error2 = np.loadtxt(file2, usecols=(0, 1, 2), delimiter='\t', unpack=True)
x3, y3, error3 = np.loadtxt(file3, usecols=(0, 1, 2), delimiter='\t', unpack=True)

popt1, pcov1 = curve_fit(fit_func, x1[1:], y1[1:], sigma=error1[1:], absolute_sigma=True)
popt2, pcov2 = curve_fit(fit_func, x2[1:], y2[1:], sigma=error2[1:], absolute_sigma=True)
popt3, pcov3 = curve_fit(fit_func, x3[1:], y3[1:], sigma=error3[1:], absolute_sigma=True)

# Plot
fig, axes = plt.subplots(1, 3, figsize=(12, 5))
viridis = plt.cm.viridis(np.linspace(0, 1, 5))
# Primo plot
axes[0].errorbar(x1, y1, yerr=error1, fmt='o', color=viridis[0], capsize=0, elinewidth=3, alpha=0.4, label='Data')
axes[0].plot(x1, fit_func(x1, *popt1), 'r-',
             label=rf'$f(x) = {popt1[0]:.4f} \sqrt{{N}}$')
axes[0].set_xlabel('Number of throws')
axes[0].set_ylabel(r'$\sqrt{|r_{RW}|^2}$')
axes[0].grid(axis = 'y')
axes[0].legend()
axes[0].text(0.95, 0.05, 'a = 1', transform=axes[0].transAxes,
             fontsize=12, verticalalignment='bottom', horizontalalignment='right',
             bbox=dict(boxstyle='round,pad=0.3', facecolor='white', alpha=0.7))

# Secondo plot
axes[1].errorbar(x2, y2, yerr=error2, fmt='o', color=viridis[1], capsize=0, elinewidth=3, alpha=0.4, label='Data')
axes[1].plot(x2, fit_func(x2, *popt2), 'r-',
             label=rf'$f(x) = {popt2[0]:.4f} \sqrt{{N}}$')
axes[1].set_xlabel('Number of throws')
axes[1].set_ylabel(r'$\sqrt{|r_{RW}|^2}$')
axes[1].grid(axis = 'y')
axes[1].legend()
axes[1].text(0.95, 0.05, 'a = 3', transform=axes[1].transAxes,
             fontsize=12, verticalalignment='bottom', horizontalalignment='right',
             bbox=dict(boxstyle='round,pad=0.3', facecolor='white', alpha=0.7))

# Terzo plot
axes[2].errorbar(x3, y3, yerr=error3, fmt='o', color=viridis[2], capsize=0, elinewidth=3, alpha=0.4, label='Data')
axes[2].plot(x3, fit_func(x3, *popt3), 'r-',
             label=rf'$f(x) = {popt3[0]:.4f} \sqrt{{N}}$')
axes[2].set_xlabel('Number of throws')
axes[2].set_ylabel(r'$\sqrt{|r_{RW}|^2}$')
axes[2].grid(axis = 'y')
axes[2].legend()
axes[2].text(0.95, 0.05, 'a = 10', transform=axes[2].transAxes,
             fontsize=12, verticalalignment='bottom', horizontalalignment='right',
             bbox=dict(boxstyle='round,pad=0.3', facecolor='white', alpha=0.7))

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Carica i dati
data_1s = np.loadtxt("Exercise_2_2/trajectory_discrete.dat")
data_2p = np.loadtxt("Exercise_2_2/trajectory_continuous.dat")

fig1 = plt.figure(figsize=(15, 10))
plt.style.use('default')

ax1 = fig1.add_subplot(121, projection='3d')

colors_1s = np.linspace(0, 1, len(data_1s))
sc1 = ax1.scatter(data_1s[:, 1], data_1s[:, 2], data_1s[:, 3],
                  c=colors_1s, cmap="viridis", s=10, alpha=0.9)

ax1.plot(data_1s[:, 1], data_1s[:, 2], data_1s[:, 3],
         color='black', linewidth=1, alpha=0.4)

ax1.set_title("Discrete Random Walk", fontsize=14)
ax1.set_xlabel("X position  [$a$]", fontsize=12)
ax1.set_ylabel("Y position  [$a$]", fontsize=12)
ax1.set_zlabel("Z position  [$a$]", fontsize=12)
ax1.view_init(elev=30, azim=45)

cbar1 = fig1.colorbar(sc1, ax=ax1, shrink=0.6)
cbar1.set_label("Time evolution", fontsize=12)

ax2 = fig1.add_subplot(122, projection='3d')

colors_2p = np.linspace(0, 1, len(data_2p))
sc2 = ax2.scatter(data_2p[:, 1], data_2p[:, 2], data_2p[:, 3],
                  c=colors_2p, cmap="viridis", s=10, alpha=0.9)

ax2.plot(data_2p[:, 1], data_2p[:, 2], data_2p[:, 3],
         color='black', linewidth=1, alpha=0.4)

ax2.set_title("Continuous Random Walk", fontsize=14)
ax2.set_xlabel("X position  [$a$]", fontsize=12)
ax2.set_ylabel("Y position  [$a$]", fontsize=12)
ax2.set_zlabel("Z position  [$a$]", fontsize=12)
ax2.view_init(elev=30, azim=45)

cbar2 = fig1.colorbar(sc2, ax=ax2, shrink=0.6)
cbar2.set_label("Time evolution", fontsize=12)

plt.tight_layout()
plt.show()
```
