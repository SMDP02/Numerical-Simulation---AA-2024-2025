```python
import matplotlib.pyplot as plt
import numpy as np

file = "output_average.data"
fig, ax = plt.subplots(figsize=(6, 5))

x, y, error = np.loadtxt(file, usecols=(0, 1, 2), delimiter='\t', unpack=True)
ax.errorbar(x, y - 0.5, yerr=error, fmt='o', color='darkcyan', capsize=0, elinewidth=2, alpha=0.4)
ax.axhline(y=0, color='red', linestyle='--', linewidth=1)
ax.set_xlabel('Number of throws')
ax.set_ylabel('Average - 0.5')
ax.grid(True)

plt.tight_layout()
plt.show()
```
